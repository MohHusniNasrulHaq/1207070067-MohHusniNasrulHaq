```python
import matplotlib.pyplot as plt #mengimport library untuk menghasilkan grafik
import numpy as np #mengimport library numpy untuk array dan matriks
import cv2 #library untuk operasi dalam citra
```

```python
img = cv2.imread('prabowo.jpeg') #membuka gambar dengan opencv
```

```python
prabowo = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # mengubah format gambar dari BGR menjadi RGB
```

```python
# Menampilkan gambar awal (citra RGB) tanpa filter menggunakan plt.imshow
kernel = np.ones((5,5),np.float32)/25
print(kernel)
# Membuat filter dengan matriks berukuran 5x5 
```

```python
# Melakukan operasi filtering pada citra asli (img) menggunakan filter yang telah dibuat sebelumnya 
prabowo_filter = cv2.filter2D(img,-1,kernel)
```

```python
# Menampilkan citra setelah proses filtering menggunakan plt.imshow
plt.imshow(prabowo_filter)
plt.title("Gambar Setelah Filtering")
plt.axis("off")
plt.show()
```

```python
#high Pass Filter
img = cv2.imread('prabowo.jpeg') #membuka gambar dengan opencv
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)  #Menggunakan operasi Laplace untuk high-pass filtering
```

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=7) #Menggunakan opersi Sobel pada sumbu X 
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=7) #Menggunakan operasi Sobel pada sumbu Y
```

```python
# Perbesar ukuran hasil plotting
plt.rcParams["figure.figsize"] = (20, 20)

# Menampilkan citra asli
plt.subplot(2, 2, 1) #ngesubplot dengan matriks 2x2 dan tampil di citra grafik1
plt.imshow(img, cmap='gray') #memunculkan hasil citra grayscale
plt.title('Original') #memberi judul
plt.xticks([])  #Menghilangkan sumbu x
plt.yticks([])  #Menghilangkan sumbu y

# Menampilkan hasil Laplacian
plt.subplot(2, 2, 2) #ngesubplot dengan matriks 2x2 dan tampil di citra grafik2
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian')  # Judul plot
plt.xticks([])  # Menghilangkan sumbu x
plt.yticks([])  # Menghilangkan sumbu y

# Menampilkan hasil Sobel X
plt.subplot(2, 2, 3) #ngesubplot dengan matriks 2x2 dan tampil di citra grafik3
plt.imshow(sobelx, cmap='gray') #memunculkan hasil citra grayscale
plt.title('Sobel X') #memberi judul
plt.xticks([]) #Menghilangkan sumbu x
plt.yticks([]) #Menghilangkan sumbu y

# Menampilkan hasil Sobel Y
plt.subplot(2, 2, 4) #ngesubplot dengan matriks 2x2 dan tampil di citra grafik4
plt.imshow(sobely, cmap='gray') #memunculkan hasil citra grayscale
plt.title('Sobel Y') #memberi judul
plt.xticks([]) #Menghilangkan sumbu x
plt.yticks([]) #Menghilangkan sumbu y

plt.show()#Menampilkan plot
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(121)
hist_data = img.flatten()
plt.hist(hist_data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram (Original)')

plt.subplot(122)
hist_data = laplacian.flatten()
plt.hist(hist_data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram (Laplacian)')

plt.show()
```

```python
#Thresholding
img = cv2.imread("prabowo.jpeg",0) # Membaca citra
```

```python
#===============Perhitungan threshold================#
# Perhatikan nilai ambang batas bawah dan atas dari tiap fungsi yang diberikan
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
```

```python
# Menampilkan hasil
titles = ['Gambar asli','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
```

```python
# Menyimpan judul dan citra hasil thresholding
for i in range(6):
    plt.subplot(3,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
# Menampilkan citra-citra hasil thresholding dalam satu plot menggunakan `plt.subplot`
```

```python
img = cv2.medianBlur(img,5) # Menggunakan median blur pada citra grayscale
```

```python
# Binary Threshold
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
# Adaptive Threshold dengan Mean
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
```

```python
# Adaptive Threshold dengan Gaussian
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
```

```python
# Menampilkan hasil
titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]
```

```python
# Menampilkan hasil
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])  # Judul plot
    plt.xticks([])  # Menghilangkan sumbu x
    plt.yticks([])  # Menghilangkan sumbu y
plt.tight_layout()
plt.show() #menampilkan seluruh citra
```

```python
# Mengatur tata letak subplotdan menampilkan plot.
# Membuat histogram dari gambar asli
plt.subplot(2, 2, 1)
plt.hist(img.ravel(), 256, [0, 256])
plt.title('Histogram Gambar asli')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
```

```python
# Membuat histogram dari gambar setelah di filter
for i in range(1, 4):
    plt.subplot(2, 2, i + 1)
    plt.hist(images[i].ravel(), 256, [0, 256])
    plt.title('Histogram ' + titles[i])
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
